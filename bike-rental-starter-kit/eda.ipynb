```python
import pandas as pd
import glob

# This pattern matches all 12 files
files = glob.glob('data/JC-2016*-citibike-tripdata.csv')


df_list = []
for filename in files:
    df_list.append(pd.read_csv(filename))

bikeDF = pd.concat(df_list, ignore_index=True)
```

```python
bikeDF.rename(columns={ 'Trip Duration': 'Trip_Duration_minutes'}, inplace=True)
display(bikeDF.head())
```

```python
bikeDF.columns = bikeDF.columns.str.replace(' ', '_').str.strip()
bikeDF['Start_Time'] = pd.to_datetime(bikeDF['Start_Time'])
bikeDF['Stop_Time'] = pd.to_datetime(bikeDF['Stop_Time'])
bikeDF = bikeDF.dropna(subset = ['User_Type'])
bikeDF = bikeDF.dropna(how='all', axis=0)
bikeDF['Gender'] = bikeDF.Gender.apply(lambda x: 'Male' if x==1 else ('Female' if x==2 else 'Unknown'))
max_outliers_duration = bikeDF[bikeDF['Trip_Duration_minutes'] >= 10080].index
bikeDF = bikeDF.drop(max_outliers_duration)
bikeDF['Start_Day'] = bikeDF['Start_Time'].dt.day_name()
bikeDF['Stop_Day'] = bikeDF['Stop_Time'].dt.day_name()
bikeDF.head()
```

```python
display(bikeDF.info())
print(bikeDF.describe())

display(bikeDF[bikeDF['Trip_Duration_minutes'] >= 10080])

```

```python
print(bikeDF.isna().sum()) #Birth_Year that is missing is 7% of the data, so we can drop it
bikeDF.dropna(subset=['Birth_Year'], inplace=True)
bikeDF.Birth_Year = bikeDF.Birth_Year.astype('Int64') 
bikeDF.columns = bikeDF.columns.str.lower()
bikeDF.to_csv('data/JC-2016-citibike-tripdata-cleaned.csv', index=False)
```

```python
from sqlalchemy import create_engine, Column, Integer, String, Float, DateTime, ForeignKey
from sqlalchemy.orm import declarative_base, relationship, sessionmaker
import pandas as pd

Base = declarative_base()

class BikeUser(Base):
    __tablename__ = 'bike_users'
    user_id = Column(Integer, primary_key=True)
    user_type = Column(String)
    birth_year = Column(Integer)
    gender = Column(String)
    
    trips = relationship("BikeTrip", back_populates="user")

class Station(Base):
    __tablename__ = 'stations'
    station_id = Column(Integer, primary_key=True)
    station_name = Column(String)
    latitude = Column(Float)
    longitude = Column(Float)
    
    start_trips = relationship("BikeTrip", back_populates="start_station", foreign_keys='BikeTrip.start_station_id')
    end_trips = relationship("BikeTrip", back_populates="end_station", foreign_keys='BikeTrip.end_station_id')

class BikeTrip(Base):
    __tablename__ = 'bike_trips'
    trip_id = Column(Integer, primary_key=True)
    bike_id = Column(Integer, nullable=False)
    trip_duration_minutes = Column(Float)
    start_time = Column(DateTime)
    stop_time = Column(DateTime)
    
    start_station_id = Column(Integer, ForeignKey('stations.station_id'))
    end_station_id = Column(Integer, ForeignKey('stations.station_id'))
    user_id = Column(Integer, ForeignKey('bike_users.user_id'))
    weather_id = Column(Integer, ForeignKey('weather.id'))
    
    start_station = relationship("Station", foreign_keys=[start_station_id], back_populates="start_trips")
    end_station = relationship("Station", foreign_keys=[end_station_id], back_populates="end_trips")
    user = relationship("BikeUser", back_populates="trips")
    weather = relationship("Weather", back_populates="trips")
class Weather(Base):
    __tablename__ = 'weather'
    id = Column(Integer, primary_key=True)
    date = Column(DateTime, nullable=False, unique=True)
    day_of_week = Column(String)
    average_daily_wind_speed_mph = Column(Float)
    precipitation_inches = Column(Float)
    snowfall_inches = Column(Float)
    snow_depth_inches = Column(Float)
    average_temperature_f = Column(Float)
    max_temperature_f = Column(Float)
    min_temperature_f = Column(Float)
    wdf2 = Column(Float)
    wdf5 = Column(Float)
    wsf2 = Column(Float)
    wsf5 = Column(Float)
    
    trips = relationship("BikeTrip", back_populates="weather")
WeatherDF = WheatherDF[['date', 'day_of_week', 'average_daily_wind_speed_mph', 'precipitation_inches', 'snowfall_inches', 'snow_depth_inches', 'average_temperature_f', 'max_temperature_f', 'min_temperature_f', 'wdf2', 'wdf5', 'wsf2', 'wsf5']]
engine = create_engine("postgresql+psycopg2://postgres:postgres@localhost:5432/CitiBike", echo=True)
Session = sessionmaker(bind=engine)
session = Session()
Base.metadata.create_all(engine)

WeatherDF.to_sql('weather', con=engine, if_exists='append', index=False)

unique_users = bikeDF[['user_type', 'birth_year', 'gender']].drop_duplicates()

user_id_map = {}
for _, row in unique_users.iterrows():
    user = BikeUser(
        user_type=row['user_type'],
        birth_year=row['birth_year'],
        gender=row['gender']
    )
    session.add(user)
    session.flush()  
    user_id_map[(row['user_type'], row['birth_year'], row['gender'])] = user.user_id

session.commit()


def get_user_id(row):
    return user_id_map.get((row['user_type'], row['birth_year'], row['gender']))

bikeDF['user_id'] = bikeDF.apply(get_user_id, axis=1)


BikeStart = bikeDF[['start_station_id', 'start_station_name', 'start_station_latitude', 'start_station_longitude']].rename(
    columns={'start_station_id': 'station_id',
             'start_station_name': 'station_name',
             'start_station_latitude': 'latitude',
             'start_station_longitude': 'longitude'})
BikeEnd = bikeDF[['end_station_id', 'end_station_name', 'end_station_latitude', 'end_station_longitude']].rename(
    columns={'end_station_id': 'station_id',
             'end_station_name': 'station_name',
             'end_station_latitude': 'latitude',
             'end_station_longitude': 'longitude'})
BikeStation = pd.concat([BikeStart, BikeEnd]).drop_duplicates().reset_index(drop=True)
BikeStation.to_sql('stations', con=engine, if_exists='append', index=False)


# Map start_time.date() to weather_id
weather_df = pd.read_sql_table('weather', con=engine)
weather_date_to_id = {row['date'].date(): row['id'] for _, row in weather_df.iterrows()}
bikeDF['weather_id'] = bikeDF['start_time'].dt.date.map(weather_date_to_id)

# Prepare and insert trips
Biketrips = bikeDF[['bike_id', 'trip_duration_minutes', 'start_time', 'stop_time', 'start_station_id', 'end_station_id', 'user_id', 'weather_id']]
Biketrips.to_sql('bike_trips', con=engine, if_exists='append', index=False)


    
```
